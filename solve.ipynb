```python
# 读取BCHAIN-MKPRU.csv文件
df_BM = pd.read_csv('BCHAIN-MKPRU.csv', encoding='utf-8')
# 读取LBMA-GOLD.csv文件
df_GOLD = pd.read_csv('LBMA-GOLD.csv', encoding='utf-8')

df = pd.merge(df_BM, df_GOLD, how='outer', on='Date')

df["Date"] = df["Date"].apply(lambda x: time.strftime('%Y/%m/%d', time.strptime(x, '%m/%d/%y')))

df = df.sort_values("Date")

df.columns=['Date','bitcoin','gold']

df['gold_future']=df['gold'].fillna(method='bfill')
df_origin=df
df_origin
```

```python

from sklearn.metrics import r2_score

def split_sequences(dataset, look_back):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back)]
        dataX.append(a)
        dataY.append(dataset[i + look_back])
    return numpy.array(dataX),numpy.array(dataY)
    
def lstm(dataset,n_steps,path):

    in_seq1= dataset[:,0].reshape((dataset.shape[0], 1))
    X, y = split_sequences(dataset, n_steps)
    model = Sequential()
    model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(n_steps, 1)))
    model.add(LSTM(50, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    model.fit(X, y, epochs=500, verbose=2,shuffle=False)
    model.save(path)
    last_input=np.array(dataset[-1:,:])
    return X,y,last_input,n_steps

```

```python
dataset = df_BM['Value'].values
# 将整型变为float
dataset = dataset.astype('float32').reshape(-1, 1) 
#对数据集合进行标准化
scaler = MinMaxScaler(feature_range=(0, 1))
dataset=scaler.fit_transform(dataset).astype('float32')
X,y,x_input_last,n_steps=lstm(dataset,1,'BM.h5')

```

```python

model=load_model('BM.h5') 
trainPredict = model.predict(X)
testPredict = scaler.inverse_transform(trainPredict)
testY = scaler.inverse_transform(y)
df_BM['Bit_T']=pd.DataFrame(testPredict[:,0],columns=['Bit_T'])
df_BM.dropna(inplace=True)
df_BM    
```

```python
from sko.operators import ranking, selection, crossover, mutation
from sko.GA import GA
```

```python
def sign(m):
    return m/abs(m+1e7)
def model_1_ga(z1,z2,m1,m2,n1,n2,us):
    demo_func = demo_func = lambda x: -((n1 / m1) * (z1 + x[0]) + (n2 / m2) * (z2 + x[1])+(us-(x[0]*(1+sign(x[0])*0.04)+x[1]*(1+sign(x[1])*0.08))))

    constraint_ueq = [lambda x: x[0]*(1+sign(x[0])*0.04)+x[1]*(1+sign(x[1])*0.08) - us,
        lambda x: (-z1 - z2) - (x[0] + x[1]),
        lambda x: 0.04  - ((n1 / m1) - 1)* sign(x[0]),
        lambda x: 0.08 - ((n2 / m2) - 1)* sign(x[1]),
        
            ]

    ga = GA(func=demo_func, n_dim=2, size_pop=100, max_iter=500, prob_mut=0.001,
            lb=[-z1, -z2], 
            ub=[us/1.04, us/1.08], 
            precision=[1e-7, 1e-7], 
            constraint_ueq=constraint_ueq)

    ga.register(operator_name='selection', operator=selection.selection_roulette_2)
    ga.register(operator_name='ranking', operator=ranking.ranking)
    ga.register(operator_name='crossover', operator=crossover.crossover_2point_bit)
    ga.register(operator_name='mutation', operator=mutation.mutation)
    best, func = ga.run()
    
    return -func[0],best[0],best[1],(us-(best[0]*(1+sign(best[0])*0.04)+best[1]*(1+sign(best[1])*0.08)))
```

```python
def model_2_ga(z1,z2,m2,n2,us):

    demo_func = lambda x: -(z1 + (n2 / m2) * (z2 + x[0])+(us-x[0]*(1+sign(x[0])*0.08)))


    constraint_ueq = [
        lambda x: x[0]*(1+sign(x[0])*0.08) - us,
        lambda x: - z2 - x[0],
        lambda x: 0.08  - ((n2 / m2) - 1)*sign(x[0])
    ]
    ga = GA(func=demo_func, n_dim=1, size_pop=100, max_iter=500, prob_mut=0.001,
            lb=[-z2], ub=[us/1.08], precision=[1e-7], constraint_ueq=constraint_ueq)

    ga.register(operator_name='selection', operator=selection.selection_roulette_2)
    ga.register(operator_name='ranking', operator=ranking.ranking)
    ga.register(operator_name='crossover', operator=crossover.crossover_2point_bit) 
    ga.register(operator_name='mutation', operator=mutation.mutation)
    best, func = ga.run()
    return -func[0],best[0],(us-best[0]*(1+sign(best[0])*0.08))
```

```python
# 黄金持有
us=1000
z1 = 0
z2 = 0
alldata=[]
for i in range(df.shape[0]):
    try:
        if i== df.shape[0]-15:
            break
        # m1 真实的今日黄金价格
        m1=df['gold'].iloc[i]
       
        # m2 真实的今日比特币价格
        m2=df['bitcoin'].iloc[i]
        
        n1= df['GOLD_T'].iloc[i+1]
        # n2 预测的明日比特币价格
        n2= df['Bit_T'].iloc[i+1]
         # n1 真实的明日黄金价格
        N1= df['gold_future'].iloc[i+1]
        # n2 真实的明日比特币价格
        N2= df['bitcoin'].iloc[i+1]
        if (np.isnan(m1)==True) or (np.isnan(n1)==True):
            _,x_2,us=model_2_ga(z1,z2,m2,n2,us)
            z2=N2 / m2 * (z2 + x_2)
            alldata.append([df['Date'].iloc[i],m1,m2,n1,n2,N1,N2,z1,z2,0,x_2,us,z1+z2+us])
        else:
            _,x_1,x_2,us=model_1_ga(z1,z2,m1,m2,n1,n2,us)
            z1=N1 / m1 * (z1 + x_1) 
            z2=N2 / m2 * (z2 + x_2)
            alldata.append([df['Date'].iloc[i],m1,m2,n1,n2,N1,N2,z1,z2,x_1,x_2,us,z1+z2+us])
        if i%100==0:
            print(z1+z2+us)
    except:
        print(df['Date'].iloc[i])
        break
```
